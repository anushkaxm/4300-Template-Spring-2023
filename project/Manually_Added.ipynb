```python
def fix_percent(x):
    last_pos = 0
    pos = x.index('%')
    new_str = ""
    while pos != -1:
        new_str += x[last_pos:pos]
        #print("pos", pos, x[last_pos:pos])
        new_str += '%'
        last_pos = pos
        pos = x.find('%', pos + 1)
    #print("pos here", last_pos)
    new_str += x[last_pos:]
        
    return new_str
```
