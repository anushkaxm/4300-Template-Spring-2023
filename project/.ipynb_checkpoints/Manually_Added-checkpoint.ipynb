```python
def add_picture_links(data):
    
    lst = [('#106 Cocktail', 'cocktail white'),
           ('10 to 7', 'martini yellow'),
           ("1880's Americana", 'short orange'),
           ("20th Century", 'yellow citrus'),
           ('3 & Tree', 'brown short'),
           ('3 Degrees', 'yellow cocktail citrus'),
           ('3AM', 'clear citrus longdrink'),
           ('50/50 Split', 'martini citrus clear'),
           ('7th Heaven', 'cocktail white citrus')
          ]
    for name, tags in lst:
        data.loc[data['drink_name'] == name, 'tags'] = tags
    return data
```
