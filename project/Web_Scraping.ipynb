```python
merged_data = pd.read_csv('merged_data.csv').drop(columns = ['Unnamed: 0'])
merged_data[pd.isna(merged_data['ingredients1'])]
indexAge = merged_data[pd.isna(merged_data['ingredients1'])].index
merged_data.drop(indexAge , inplace=True)
merged_data = merged_data[merged_data['instructions'].str.contains('%') == False]
```

```python
def create_tags(category, taste):
    cat_list = set(str(category).split(","))
    taste_list = set(str(taste).split(","))
    cat_set = {i.strip() for i in cat_list}
    taste_set = {i.strip() for i in taste_list}
    if pd.isna(category) == False and pd.isna(taste) == False:
        return cat_set.union(taste_set)
    if pd.isna(taste):
        return cat_set
    if pd.isna(category):
        return taste_set
    return set()
merged_data['tags'] = merged_data.apply(lambda row: create_tags(row['category'], row['taste']), axis=1)
merged_data = merged_data.drop(columns=['category', 'taste'])
tags_column = merged_data.pop('tags')
  
merged_data.insert(3, 'tags', tags_column)
merged_data
```

```python
def format_inst(x):
    if '*' in x:
        pos = x.find('*')
        x = x[:pos]
    return x
merged_data['instructions'] = merged_data['instructions'].apply(format_inst)
```

```python
def calculate_difficulty(x):
    val = 0
    if '.' in x:
        val = x.count('.')
    elif ',' in x:
        val = x.count(',')
    return val
```

```python
merged_data['steps'] = merged_data['instructions'].apply(calculate_difficulty)
steps_column = merged_data.pop('steps')
merged_data.insert(2, 'steps', steps_column)
merged_data
```

```python
def parse_drink_name (x):
    capitals = [idx for idx, chr in enumerate(x) if chr.isupper()]
    spaces = [idx for idx, chr in enumerate(x) if chr == ' ']
    odd_names = ['TBT', 'B52', 'B54']
    
    name_splitting = x[:len(x)//2].rfind(" ")
    for cap in range(len(capitals)):
        if capitals[cap] > name_splitting:
            cap += 1
            break
    drink_name = x[:capitals[cap]]
    name_split_pos = capitals[cap]
    
    for name in odd_names:
        if name in x:
            drink_name = name
            name_split_pos = x.find(name) + len(name)
            

    pic_str = 'https://images.cocktailflow.com/v1/cocktail/w_300,h_540/cocktail_'#blueberry_ginger_bellini-1.png
    split_name = drink_name.split(" ")
    pic_str += split_name[0].lower()
    include_smoothie = '-' in drink_name and 'Smoothie' in drink_name
    for word in split_name[1:]:
        if word.isdigit():
            pic_str += word
        elif word != 'Smoothie' or include_smoothie:
            pic_str += ("_" + word.lower())
    pic_str = pic_str.replace('-', '_')
    pic_str += '-1.png'

    tags = x[name_split_pos:]
    tags_pos = [idx for idx, chr in enumerate(tags) if chr.isupper()]
    tags_list = []
    tags_list = re.findall('[A-Z][^A-Z]*', tags)
    tags_set = set([x.lower() for x in tags_list] )
        
    return (drink_name, pic_str, tags_set)

#parse_drink_name("Red RoyalWhiskeyShooterExtreme Strong")
parse_drink_name("Blueberry Almond SmoothieJuiceSmoothieNonalcoholic")
#parse_drink_name("Banana-Oat SmoothieJuiceSmoothieNonalcoholic")
#Red RoyalWhiskeyShooterExtreme Strong
```

```python
temp_df = pd.read_csv('sparkling.csv').iloc[:,[2,4]]
temp_df.columns = ['drink_name', 'drink_info']
def parse_drink_info(df, name):
    info = df[df['drink_name'] == name]['drink_info']
    ingredients = []
    quantities = []
    instructions = ""
    giving_ingredients = True
    steps = 0
    measurements = ['parts', 'scoops', 'bottles', 'quarts', 'cubes', 'litres', 'drops', 'pieces', 'cups', 'pints', 'pint', 'dash', 'part', 'piece', 'tbsp', 
                    'sugar cube', 'cup', 'bottle', 'quart', 'tsp', 'splash', 'pinch', 'drop', 'scoop']
    for row in info:
        if pd.isna(row) == False:
            if not giving_ingredients:
                instructions += (row + ". ")
                steps += 1
            if row == 'How to Mix':
                giving_ingredients = False
            elif giving_ingredients and row != 'Ingredients':
                pos = 0
                for measure in measurements:
                    if row.find(measure) != -1:
                        if measure == 'sugar cube':
                            pos = row.find(measure)
                        else:
                            pos = row.find(measure) + len(measure)
                        break
                ingredients.append((row[:pos]).strip().lower())
                quantities.append((row[pos:]).strip().lower())
    return(ingredients, quantities, instructions, steps)
parse_drink_info(temp_df, "Happy YouthChampagneSparklingMedium")
```

```python
def scrape_data(old_df, added_tag = None):
    df = pd.DataFrame(columns = merged_data.columns)
    drink_names = old_df['drink_name'].unique()
    for name in drink_names:
        drink_name, pic_str, tags_set = parse_drink_name(name)
        if added_tag:
            tags_set.add(added_tag)
        ingredients, quantities, instructions, steps = parse_drink_info(old_df, name)
        new_row = [drink_name, instructions, steps, pic_str, tags_set]
        for pos in range(len(ingredients)):
            new_row.append(quantities[pos])
            new_row.append(ingredients[pos])
        while len(new_row) < len(merged_data.columns):
            new_row.append(None)
        df.loc[len(df.index)] = new_row
    return df  
```

```python
csv_names = ['mocktails.csv', 'weak-cocktails.csv', 'light-cocktails.csv', 'medium-cocktails.csv', 
             'strong-cocktails.csv', 'extra-strong-cocktails.csv', 'sparkling.csv', 'classic.csv',
            'creamy.csv', 'frozen.csv', 'hot.csv', 'longdrink.csv', 'martini.csv', 'nonalcoholic.csv',
            'shooter.csv', 'short.csv', 'tropical.csv']
special_names = ['staying_in_by_the_fireplace.csv', 'poolside.csv', 'company_party.csv', 'downtown.csv',
               'family_gathering.csv', 'girls_night_out.csv', 'guys_night_out.csv', 'house_party.csv',
               'club.csv', 'restaurant.csv', 'movie_night.csv', 'beach.csv', 'picnic.csv', 'time_for_you.csv',
                'black.csv', 'brown.csv', 'purple.csv', 'blue.csv', 'green.csv', 'yellow.csv', 'orange.csv',
                'red.csv', 'pink.csv', 'multicolor.csv', 'white.csv', 'transparent.csv']
scraped_df = pd.DataFrame()
for file in csv_names:
    temp_df = pd.read_csv(file).iloc[:,[2,4]]
    temp_df.columns = ['drink_name', 'drink_info']
    new_data = scrape_data(temp_df)
    if len(scraped_df) >0:
        scraped_df = scraped_df.append(new_data)
    else:
        scraped_df = new_data
        
for file in special_names:
    temp_df = pd.read_csv(file).iloc[:,[2,4]]
    temp_df.columns = ['drink_name', 'drink_info']
    pos = file.find('.')
    additional_tag = file[:pos].replace("_", " ")
    new_data = scrape_data(temp_df, additional_tag)
    scraped_df = scraped_df.append(new_data)
```

```python
def agg_tags(x):
    new_set = set()
    for lst in x:
        if pd.isna(lst) == False:
            new_set = new_set.union(lst)
    return new_set
```

```python
def edit_ingr(x):
    extra_chars = ["'", " ", "*"]
    if pd.isna(x) == False:
        length = len(x)
        same = False
        while same == False:
            for char in extra_chars:
                x = x.strip(char)
            if len(x) == length:
                same = True
            length = len(x)
    return x
```

```python
together['ingredients1'] = together['ingredients1'].apply(edit_ingr)
```

```python
ingr_cols = ['ingredients1','ingredients2','ingredients3','ingredients4','ingredients5','ingredients6',
   'ingredients7','ingredients8','ingredients9','ingredients10','ingredients11','ingredients12']
for col in ingr_cols:
    together[col] = together[col].apply(edit_ingr)
```

```python
def list_to_str(x):
    string = ""
    for tag in x:
        string += (tag + " ")
    if x == {'nan'}:
        return None
    return string[:-1]
```

```python
together['tags'] = together['tags'].apply(list_to_str)
together
```
