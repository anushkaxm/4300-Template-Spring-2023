```python
columns = ['strDrink', 'strAlcoholic','strCategory', 'strDrinkThumb', 'strGlass', 
           'strIngredient1', 'strMeasure1', 'strIngredient2', 'strMeasure2','strIngredient3', 'strMeasure3',
           'strIngredient4', 'strMeasure4', 'strIngredient5', 'strMeasure5', 'strIngredient6', 'strMeasure6', 
           'strIngredient7', 'strMeasure7', 'strIngredient8', 'strMeasure8','strIngredient9', 'strMeasure9',
           'strIngredient10', 'strMeasure10', 'strIngredient11', 'strMeasure11','strIngredient12', 'strMeasure12']
```

```python
df = drinks[columns]
df = df.replace(np.nan, '', regex=True)
```

```python
df
```

```python
def replace_n(x):
    if x == "":
        return ""
    else:
        pos = x.find('\n')
        if pos != -1:
            return x[:pos]
    return x
```

```python
for col in columns:
    df[col] = df[col].apply(replace_n)
```

```python
df
```

```python
def count_ingredients(x, col_names):
    count = 0
    for col in col_names:
        if x[col] != "":
            count += 1
        else:
            return count
    return count
```

```python
ingr_cols = ['strIngredient1', 'strIngredient2', 'strIngredient3','strIngredient4', 'strIngredient5', 
                'strIngredient6', 'strIngredient7', 'strIngredient8','strIngredient9', 'strIngredient10', 
                'strIngredient11','strIngredient12']
df['ingr_count'] = df.apply(lambda row : count_ingredients(row, ingr_cols), axis = 1)
```

```python
df
```

```python
def clean_alcoholic(x):
    if x == "":
        return 'Alcoholic'
    if x == 'Non Alcoholic':
        return 'Non alcoholic'
    return x
df['strAlcoholic'] = df['strAlcoholic'].apply(clean_alcoholic)
```

```python
def clean_glass(x):
    if x == 'Beer Glass' or x == 'Beer mug' or x == 'Beer pilsner': return 'Beer Glass'
    if x == 'Champagne Flute' or x == 'Champagne flute': return 'Champagne Flute'
    if x == 'Cocktail Glass' or x == 'Cocktail glass': return 'Cocktail Glass'
    if x == 'Coffee Mug' or x == 'Coffee mug': return 'Coffee Mug'
    if x == 'Collins Glass' or x == 'Collins glass': return 'Collins Glass'
    if x == 'Highball Glass' or x == 'Highball glass': return 'Highball Glass'
    if x == 'Margarita glass' or x == 'Margarita/Coupette glass': return 'Margarita Glass'
    if x == 'Old-Fashioned glass' or x == 'Old-fashioned glass': return 'Old-Fashioned Glass'
    if x == 'Punch Bowl' or x == 'Punch bowl': return 'Punch Bowl'
    if x == 'Shot Glass' or x == 'Shot glass': return 'Shot Glass'
    if x == 'White wine glass' or x == 'Wine Glass': return 'Wine Glass'
    return x
df['strGlass'] = df['strGlass'].apply(clean_glass)
```

```python
dfc = df.groupby('strCategory')['strCategory'].count()

ax = dfc.plot.barh(figsize=(10, 5), xlabel='Drink Category', ylabel = 'Number of Recipies')
plt.show()
```

```python
ingr_counts = {}
ingr_cols
for col in ingr_cols:
    for row in range(len(df)):
        if df[col][row] != "":
            if df[col][row] in ingr_counts: ingr_counts[df[col][row]] += 1
            else: ingr_counts[df[col][row]] = 1
top_ingr = sorted(ingr_counts.items(), key = lambda x: x[1], reverse = True)[:10]
```
